## 1. What is Machine Learning?

**Machine Learning** is a subset of Artificial Intelligence (AI) that enables computers to **learn from data** without being explicitly programmed.

### Real-World Examples:
- **Email Spam Filter**: Learns to identify spam emails based on patterns
- **Netflix Recommendations**: Suggests movies based on your viewing history
- **Voice Assistants**: Understands and responds to your voice commands
- **Self-Driving Cars**: Makes driving decisions based on sensor data

### The Key Idea:
Instead of writing rules manually, we provide **data** and let the algorithm discover patterns!

## 2. Types of Machine Learning

### A. Supervised Learning
- **What**: Learning from labeled data (input + known output)
- **Goal**: Predict output for new, unseen inputs
- **Types**:
  - **Regression**: Predict continuous values (e.g., house prices)
  - **Classification**: Predict categories (e.g., spam or not spam)

### B. Unsupervised Learning
- **What**: Learning from unlabeled data (input only)
- **Goal**: Discover hidden patterns or groupings
- **Types**:
  - **Clustering**: Group similar items (e.g., customer segments)
  - **Dimensionality Reduction**: Reduce data complexity

### C. Reinforcement Learning
- **What**: Learning through trial and error
- **Goal**: Maximize rewards through actions
- **Example**: Game-playing AI, robotics

## 4. The Machine Learning Workflow

Every ML project follows these steps:

```
1. Define Problem     -> What are we trying to predict/discover?
        |
2. Collect Data       -> Gather relevant data
        |
3. Explore Data (EDA) -> Understand patterns, missing values, distributions
        |
4. Preprocess Data    -> Clean, transform, scale data
        |
5. Split Data         -> Training set + Test set
        |
6. Select Model       -> Choose appropriate algorithm
        |
7. Train Model        -> Fit model on training data
        |
8. Evaluate Model     -> Test performance on test data
        |
9. Tune Model         -> Improve with hyperparameter tuning
        |
10. Deploy Model      -> Put model into production
```

## 6. Exploratory Data Analysis (EDA) Basics

Before building any model, we must understand our data!

````python
# Let's explore the Iris dataset

# Basic information
print("=" * 50)
print("DATASET OVERVIEW")
print("=" * 50)
print(f"\nShape: {iris_df.shape[0]} rows, {iris_df.shape[1]} columns")
print(f"\nColumn Names: {list(iris_df.columns)}")
````

````python
# Data types and missing values
print("\nData Types:")
print(iris_df.dtypes)
print("\nMissing Values:")
print(iris_df.isnull().sum())
````

````python
# Statistical summary
print("\nStatistical Summary:")
iris_df.describe()
````

````python
# Visualization: Distribution of features
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

for idx, column in enumerate(iris.feature_names):
    ax = axes[idx // 2, idx % 2]
    for species in iris.target_names:
        data = iris_df[iris_df['species'] == species][column]
        ax.hist(data, alpha=0.7, label=species, bins=15)
    ax.set_xlabel(column)
    ax.set_ylabel('Frequency')
    ax.legend()
    ax.set_title(f'Distribution of {column}')

plt.tight_layout()
plt.show()
````

````python
# Correlation Heatmap
plt.figure(figsize=(10, 8))
correlation_matrix = iris_df.drop(['target', 'species'], axis=1).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Feature Correlation Heatmap')
plt.show()
````

````python
# Pairplot - See relationships between all features
sns.pairplot(iris_df, hue='species', diag_kind='kde')
plt.suptitle('Iris Dataset Pairplot', y=1.02)
plt.show()
````

## 7. Data Preprocessing Essentials

Before feeding data to ML models, we often need to preprocess it.

### 7.1 Train-Test Split

**Why?** We need separate data to:
- **Train** the model (learning)
- **Test** the model (evaluation)

````python
# Prepare features (X) and target (y)
X = iris_df.drop(['target', 'species'], axis=1)
y = iris_df['target']

# Split: 80% training, 20% testing
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2,      # 20% for testing
    random_state=42,    # For reproducibility
    stratify=y          # Maintain class distribution
)

print(f"Training set size: {len(X_train)}")
print(f"Test set size: {len(X_test)}")
print(f"\nTraining target distribution:\n{y_train.value_counts()}")
````

### 7.2 Feature Scaling

**Why?** Many ML algorithms perform better when features are on similar scales.

Common methods:
- **StandardScaler**: Mean=0, Std=1
- **MinMaxScaler**: Range [0, 1]

````python
# StandardScaler example
scaler = StandardScaler()

# Fit on training data, transform both
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Compare before and after
print("Before Scaling:")
print(f"Mean: {X_train.mean().values}")
print(f"Std: {X_train.std().values}")

print("\nAfter Scaling:")
print(f"Mean: {X_train_scaled.mean(axis=0).round(2)}")
print(f"Std: {X_train_scaled.std(axis=0).round(2)}")
````

## 8. Your First ML Model!

Let's build a simple classification model using K-Nearest Neighbors (KNN).

````python
from sklearn.neighbors import KNeighborsClassifier

# Step 1: Create the model
knn_model = KNeighborsClassifier(n_neighbors=3)

# Step 2: Train the model
knn_model.fit(X_train_scaled, y_train)

# Step 3: Make predictions
y_pred = knn_model.predict(X_test_scaled)

# Step 4: Evaluate
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy * 100:.2f}%")
````

````python
# Detailed evaluation with confusion matrix
from sklearn.metrics import classification_report, confusion_matrix

print("Confusion Matrix:")
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=iris.target_names,
            yticklabels=iris.target_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=iris.target_names))
````
