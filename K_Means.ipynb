### K-Means Clustering Example

K-means is a popular unsupervised machine learning algorithm used for partitioning `n` observations into `k` clusters, where each observation belongs to the cluster with the nearest mean (cluster centroid), serving as a prototype of the cluster.

```python
# Import necessary libraries
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
```

First, let's generate some synthetic data that has distinct clusters. We'll use `make_blobs` for this purpose.

```python
# Generate synthetic data
n_samples = 300
n_features = 2
n_clusters = 3 # We know there are 3 clusters in our synthetic data

X, y = make_blobs(n_samples=n_samples, n_features=n_features, centers=n_clusters, random_state=42)

print(f"Shape of generated data (X): {X.shape}")
print(f"Shape of true labels (y): {y.shape}")

# Visualize the generated data (without knowing the clusters yet)
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], s=50, cmap='viridis')
plt.title('Generated Data (Before Clustering)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.show()
```

Now, let's apply the K-means algorithm to this data. We need to specify the number of clusters (`n_clusters`).

```python
# Initialize KMeans with the desired number of clusters
kmeans = KMeans(n_clusters=n_clusters, init='k-means++', max_iter=300, n_init=10, random_state=42)

# Fit the model to the data
kmeans.fit(X)

# Get the cluster assignments for each data point
labels = kmeans.labels_

# Get the coordinates of the cluster centers
centers = kmeans.cluster_centers_

print("Cluster labels for first 10 data points:", labels[:10])
print("Cluster centers:\n", centers)
```

Finally, let's visualize the results, showing the data points colored by their assigned cluster and marking the cluster centers.

```python
# Visualize the clustered data
plt.figure(figsize=(8, 6))

# Plot data points colored by cluster assignment
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis', alpha=0.7)

# Plot the cluster centers
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.9, marker='X', label='Cluster Centers')

plt.title(f'K-Means Clustering with {n_clusters} Clusters')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```
